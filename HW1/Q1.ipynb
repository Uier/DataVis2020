```python
import pandas as pd
import numpy as np
```

```python
# Read the data into a data frame.
_df = pd.DataFrame( pd.read_csv('pitching.csv') )
_df
```

```python
# Display all pitchers in 2015 under each team sorted by their last names.
df = _df[_df['yearID'] == 2015]
df.sort_values(['teamID', 'nameLast']).set_index(keys=['teamID', 'playerID'])[['nameFirst', 'nameLast', 'nameGiven']]
# df = df.sort_values(['teamID', 'nameLast'])
# df = df.set_index(keys=['teamID', 'playerID'])
# df[['nameFirst', 'nameLast', 'nameGiven']]
```

```python
# Display all pitchers in 2015 sorted by their ERAs. This allows you to identify the best and worst pitchers.
df = _df[_df['yearID'] == 2015]
df.set_index(keys=['playerID'], inplace=True)
df.sort_values(by='ERA')[['ERA'] + [col for col in df.columns if col is not 'ERA']]
```

```python
# Display all pitchers in 2015 under each team sorted by their ERAs.
df = _df[_df['yearID'] == 2015]
df.sort_values(['teamID', 'ERA']).set_index(keys=['teamID', 'playerID'])[['ERA', 'nameFirst', 'nameLast', 'nameGiven']]
# df = df.sort_values(['teamID', 'ERA'])
# df = df.set_index(keys=['teamID', 'playerID'])
# df[['ERA', 'nameFirst', 'nameLast', 'nameGiven']]
```

```python
# Display all pitchers in 2015 under each team sorted by their strike outs (SO).
df = _df[_df['yearID'] == 2015]
df.sort_values(['teamID', 'SO']).set_index(keys=['teamID', 'playerID'])[['SO', 'nameFirst', 'nameLast', 'nameGiven']]
# df = df.sort_values(['teamID', 'SO'])
# df = df.set_index(keys=['teamID', 'playerID'])
# df[['SO', 'nameFirst', 'nameLast', 'nameGiven']]
```

```python
# Calculate the average ERA for each team, list the teams by the average ERA in a descending order.
_df.groupby('teamID')['ERA'].mean().to_frame().sort_values(by='ERA', ascending=False)
```

```python
# Calculate the total number of home runs allowed (HRs) for each team, and list the team by the count in an ascending order.
_df.groupby('teamID')['HR'].count().to_frame().sort_values(by='HR')
```

```python
# Calculate each teamâ€™s average ERA from 2011 to 2015, and list the result as a table, where each column is a year, each row is a team.
pd.DataFrame({f'{year} ERA': _df[_df['yearID'] == 2011].groupby('teamID')['ERA'].mean() for year in range(2011,2016)})
```

```python
# Calculate the average ERAs for American League (AL) and National League (NL) from 2011 to 2015, and list the result as a table, where each column is a league and each row is a year.
lg = {'AL': [], 'NL': []}
idx = [y for y in range(2011, 2016)]
for year in idx:
    for c in ['AL', 'NL']:
        lg[c].append(_df[_df['yearID'] == year].groupby('lgID')['ERA'].mean()[c])
pd.DataFrame(lg, idx, columns=['AL', 'NL'])
```

```python
# List the pitchers who had the most wins in each of the years from 2011 to 2015.
pitch = {'playerID': [], 'W': []}
idx = [y for y in range(2011, 2016)]
for year in idx:
    mx = _df[_df['yearID'] == year]['W'].max()
    pitch['W'].append(mx)
    pitch['playerID'].append(*_df[(_df['yearID'] == year) & (_df['W'] == mx)][['playerID']].values[0])
pd.DataFrame(pitch, idx, columns=['playerID', 'W'])
```
